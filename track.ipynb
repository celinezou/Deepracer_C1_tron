```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.load("reInvent2019_wide_cw.npy")
```

```python
def plot_points(ax, points):
    ax.scatter(points[:-1,0], points[:-1,1], s=1)
    for i,p in enumerate(points):
        ax.annotate(i, (p[0], p[1]))
```

```python
# Plot the results
fig, ax = plt.subplots(figsize=(20,10))
plot_points(ax, a[:-1,0:2])
plot_points(ax, a[:-1,2:4])
plot_points(ax, a[:-1,4:6])
ax.axis('equal')
plt.show()
```
